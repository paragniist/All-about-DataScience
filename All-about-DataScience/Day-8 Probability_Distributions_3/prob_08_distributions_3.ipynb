```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom
```

# Poisson 

```python
mu = 18.5
x_values = np.arange(0, 30)
y_values = poisson.pmf(x_values, mu=mu)
sns.barplot(x=x_values, y=y_values)
plt.show()
```

```python
def geometric():
    val = np.random.choice(["s", "f"], p=[1/6, 5/6])
    if val == "s":
        return 1
    else:
        return 1 + geometric()
    
```

```python
np.mean([geometric() for i in range(100000)])
```
