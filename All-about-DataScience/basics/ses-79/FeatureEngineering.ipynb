```python
import pandas as pd
```

```python
d = {
    "Company" : ["Google", "Microsoft", "Amazon","Linkedin"],
    "Role" : ["data scientist", "Manager", "HR","HR"]
}

df = pd.DataFrame(d)
print(df)
```

 > using label_encoder to convert categorical data into numerical data.
 Creating a new feature from an already existing feature like (compony_n from compony)

```python
from sklearn.preprocessing import LabelEncoder
import sklearn.preprocessing as abc
label_encoder = LabelEncoder()
df["Company_n"] = label_encoder.fit_transform(df['Company'])
df["Role_n"] = label_encoder.fit_transform(df['Role'])
print(df)
```

This code converts text (categorical) columns to numbers, which is required for most machine learning algorithms.​

Quick Explanation
from sklearn.preprocessing import LabelEncoder: Imports the LabelEncoder class from scikit-learn for converting categories to numbers.​

label_encoder = LabelEncoder(): Creates a LabelEncoder object.​

df["Company_n"] = label_encoder.fit_transform(df['Company']): Changes the 'Company' column’s unique values into integers and stores the result in a new column, 'Company_n'.​

df["Role_n"] = label_encoder.fit_transform(df['Role']): Applies the same process to the 'Role' column and makes 'Role_n'.​

print(df): Prints the DataFrame to show the original and new numeric columns.

Result:
The DataFrame will now have original columns and the new numeric columns, replacing categories like names or roles with numbers. This step is standard in ML preprocessing to handle non-numeric data.

> LabelEncoder is a predefinded Sk-learn class that we just need to import.
