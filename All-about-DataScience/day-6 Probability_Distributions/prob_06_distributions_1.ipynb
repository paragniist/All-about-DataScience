A bag has 3 red and 2 blue balls.


You pick a ball, write its colour, and put it back in the bag. 

This is done 4 times in total.

If all 4 times, the red ball was drawn, you win Rs 150, otherwise you lose Rs 10.

Would you play this game?

```python
rolls = np.random.choice(["R", "R", "R", "B", "B"], size=4)
rolls
```

```python
np.count_nonzero(rolls=="R")
pd.value_counts(rolls)
```

```python
red_values = []

for person in range(10000):
    rolls = np.random.choice(["R", "R", "R", "B", "B"], size=4)
    num_reds = np.count_nonzero(rolls=="R")
    red_values.append(num_reds)

pd.value_counts(red_values, normalize=True)
```

```python
np.mean(red_values)
```

## Rule-based probability 

PMF: Probability Mass Function

```python
math.comb(4, 0) # 4C0
```

```python
math.comb(4, 0) * (3/5)**0 * (2/5)**4 
```

```python
binom.pmf(n=4, p=3/5, k=0)
```

```python
math.comb(4, 1) * (3/5)**1 * (2/5)**3
```

```python
binom.pmf(n=4, p=3/5, k=1)
```

```python
math.comb(4, 2) * (3/5)**2 * (2/5)**2
```

```python
binom.pmf(n=4, p=3/5, k=2)
```

```python
math.comb(4, 3) * (3/5)**3 * (2/5)**1
```

```python
binom.pmf(n=4, p=3/5, k=3)
```

```python
math.comb(4, 4) * (3/5)**4 * (2/5)**0
```

```python
n = 4
p = 3/5
expectation = 0
for k in range(n+1):
    expectation += k * binom.pmf(n=n, k=k, p=p)


expectation
```
