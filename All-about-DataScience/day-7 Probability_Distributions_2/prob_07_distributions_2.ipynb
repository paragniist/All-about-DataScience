# ZScore

```python
# norm.cdf essentially is the Z-table
# input is the z-score, output is the cdf
norm.cdf(1.64)
```

Fraction of people shorter than 67.5 --> z-score of 67.5 = 1

```python
norm.cdf(1)
```

Fraction of people shorter than 62.5 --> z-score of 62.5 = -1

```python
norm.cdf(-1)
```

Frac of people within 62.5 and 67.5 -> within plus 1 std dev and minu 1 std dev

```python
norm.cdf(1) - norm.cdf(-1)
```

```python
norm.cdf(2) - norm.cdf(-2)
```

```python
norm.cdf(3) - norm.cdf(-3)
```

```python
def zscore_frac_shorter(height, mu=65, sigma=2.5):
    zscore = (height - mu)/sigma
    frac_shorter = norm.cdf(zscore)
    print(height, zscore, frac_shorter)
```

```python
zscore_frac_shorter(67)
```

```python
zscore_frac_shorter(69)
```

```python
zscore_frac_shorter(66)
```

```python
zscore_frac_shorter(65.748)
```

```python
zscore_frac_shorter(69.1)
```

```python
zscore_frac_shorter(65.3)
```

```python
z_values = np.linspace(-3, 3, 1000)
y_values = norm.cdf(z_values)
```

```python
plt.plot(z_values, y_values)
```
