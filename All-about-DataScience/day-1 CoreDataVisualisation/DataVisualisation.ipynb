## Univariate Analysis :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('./final_vg.csv')
data
```

```python
cat_counts = data["Genre"].value_counts() ## Categorical data of types of genre in data
cat_counts
```

```python
plt.figure(figsize=(15,8))
x = cat_counts.index
y=cat_counts
plt.bar(x,y)
```

```python
x = cat_counts.index
y=cat_counts
plt.bar(x,y,width=0.5,color="orange")
plt.xticks(rotation=90,fontsize="15")
plt.ylabel("Count")
plt.xlabel("Games Per Genre")
plt.show()
```

```python
## Seaborn is more efficient and advance than matplotlib

sns.countplot(
x=data["Genre"],
color="blue")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(
x=data["Platform"],
color="blue")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=data,
y="Genre",
order=data["Genre"].value_counts().index,
color="blue")
```

```python
# Numerical/continuos
plt.hist(data["Year"])
plt.show()
```

```python
plt.hist(data["Year"],bins=4)
plt.show()
```

```python
count,bins,_= plt.hist(data["Year"],bins=10)
```

```python
# Seaborn is more easy and effiicient
sns.histplot(data["Year"],bins=4)
```

```python
sns.histplot(data=data,
x="Year",
bins=40)
```

```python
sns.kdeplot(x=data["Year"])
## kernel-density plot. KDE-plot is not in matplot only sns
```

Kernel Density Estimate (KDE) Plot
A KDE plot, similar to histrogram, is a method for visualizing the distributions

But instead of bars, KDE represents data using a continuous probability density curve

Now, Why do we even need KDE plots?
Compared to histogram, KDE produces a plot which is less cluttered and more interpretable
Think of it as a smoothened version of histogram
