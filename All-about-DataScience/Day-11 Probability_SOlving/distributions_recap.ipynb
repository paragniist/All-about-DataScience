```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson # discrete
from scipy.stats import expon, norm, lognorm # continuous
```

# Distributions

## Bernoulli

```python
p = 0.7
x_values = [-1, 0, 1]
y_values = bernoulli.pmf(x_values, p=p)
sns.barplot(x=x_values, y=y_values)
plt.show()
```

```python
bernoulli(p=p).mean()
```

```python
p
```

```python
bernoulli(p=p).var()
```

```python
p*(1-p)
```

## Binomial

```python
n = 10
p = 0.7
x_values = np.arange(0, n+1)
y_values = binom.pmf(x_values, n=n, p=p)
sns.barplot(x=x_values, y=y_values)
plt.show()
```

```python
binom(n=n, p=p).mean()
```

```python
binom(n=n, p=p).var()
```

```python
binom(n=n, p=p).cdf(5)
```

## Geom

```python
p = 0.7
x_values = np.arange(0, 20)
y_values = geom.pmf(x_values, p=p)
sns.barplot(x=x_values, y=y_values)
plt.show()
```

```python
geom(p=p).mean()
```

```python
1/p
```

```python
geom(p=p).var()
```

```python
(1-p)/(p**2)
```

```python
geom(p=p).cdf(3)
```

## Poisson

```python
mu = 2.5 # rate or "lambda"
x_values = np.arange(0, 20)
y_values = poisson.pmf(x_values, mu=mu)
sns.barplot(x=x_values, y=y_values)
plt.show()
```

```python
poisson(mu=mu).mean()
```

```python
mu
```

```python
poisson(mu=mu).var()
```

```python
mu
```

```python
poisson(mu=mu).cdf(3)
```

```python
poisson(mu=mu).pmf(3)
```

## Expon

```python
mu = 2.5 # rate or "lambda"
x_values = np.linspace(0, 20, 100)
y_values = expon.pdf(x_values, scale=1/mu)
plt.plot(x_values, y_values)
plt.show()
```

```python
expon(scale=1/mu).mean()
```

```python
1/mu
```

```python
expon(scale=1/mu).var()
```

```python
1/mu**2
```

```python
expon(scale=1/mu).cdf(1.85)
```

## Norm

```python
mu = 3
sigma = 2
x_values = np.linspace(mu-3*sigma, mu+3*sigma, 1000)
y_values = norm.pdf(x_values, loc=mu, scale=sigma)
plt.plot(x_values, y_values)
plt.show()
```

```python
norm(loc=mu, scale=sigma).mean()
```

```python
mu
```

```python
norm(loc=mu, scale=sigma).var()
```

```python
sigma**2
```

```python
norm(loc=mu, scale=sigma).cdf(5)
```

## LogNorm


```python
mu = 3
sigma = 2
x_values = np.linspace(0, mu+5*sigma, 1000)
y_values = lognorm.pdf(x_values, s=sigma, scale=np.exp(mu))
plt.plot(x_values, y_values)
plt.show()
```

```python
lognorm(s=sigma, scale=np.exp(mu)).mean()
```

```python
np.exp(mu+sigma**2/2)
```

```python
lognorm(s=sigma, scale=np.exp(mu)).var()
```

```python
(np.exp(sigma**2) - 1)*np.exp(2*mu + sigma**2)
```

```python
lognorm(s=sigma, scale=np.exp(mu)).cdf(10)
```

```python
norm(loc=mu, scale=sigma).cdf(np.log(10))
```
