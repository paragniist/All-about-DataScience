> Problem 1 : Dealing with the Comma seperated values in 'Cast' column.

```python
constraint1 = df['director'].apply(lambda x: str(x).split(', ')).tolist()

## .apply is applying this split() function on all the values of the whole row.
```

* create DF from dict/list

* stack → flatten

* reset index

* rename column

* drop a useless column

```python
df_new1 = pd.DataFrame(constraint1,index=df['title'])
df_new1=df_new1.stack()
df_new1 = pd.DataFrame(df_new1.reset_index())
df_new1.head(10)
```

```python
df_new2 = pd.DataFrame(constraint1, index=df['title'])
df_new2 = df_new2.stack()
df_new2 = pd.DataFrame(df_new2.reset_index())
# Use the return value of rename() or call rename() with inplace=True but don't assign the return
df_new2 = df_new2.rename(columns={0: 'Actors'})
df_new2.drop(['level_1'], axis=1, inplace=True)
df_new2.head(25)
```

```python
## alternate and cleaner way to do the same thing:
df_exploded = pd.DataFrame({
    "title": df["title"],
    "Actors": constraint1
}).explode("Actors")
df_exploded.head(5)
```
