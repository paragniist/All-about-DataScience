```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import norm
```

```python
df = pd.read_csv("Sachin_ODI.csv")
```

```python
df.head()
```

```python
mu = df["runs"].mean()
mu
```

```python
sigma = df["runs"].std()
sigma
```

```python
df.groupby(["Inns"])["runs"].mean()
```

```python
df.groupby(["Won"])["runs"].mean()
```

```python
mu
```

```python
first_innings_runs = df[df["Inns"]==1][["runs"]]
```

```python
len(first_innings_runs)
```

```python
data = first_innings_runs
```

```python
data.mean()
```

```python
z_stat = (data.mean() - mu) / (sigma / np.sqrt(len(data)))
z_stat
```

```python
p_value = 1 - norm.cdf(z_stat)
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypotheis")
else:
    print("Stick to the null hypothesis")
```
